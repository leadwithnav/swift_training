## 🎯 Overview

So far, we've used the built-in console producers. In the real world, applications use language-specific Kafka clients. 
In this lab, you will run a custom Python script (`producer.py`) that reads e-commerce data from `orders.csv` and publishes it to Kafka using the `confluent-kafka` library.

We will use the standard single-node cluster from the root directory for this lab.

## <span style="color: #FF6B6B;">Step 1:</span> Start the Kafka Cluster


```python
!docker-compose -f ../../docker-compose.yml up -d
```

---

## <span style="color: #FF6B6B;">Step 2:</span> Create the Topic

Let's create the `orders` topic where our Python script will produce the data.

```python
!docker exec kafka kafka-topics \
  --bootstrap-server localhost:9092 \
  --create \
  --topic orders \
  --partitions 1 \
  --replication-factor 1
```

---

## <span style="color: #FF6B6B;">Step 3:</span> Install Kafka Python Client

We need to install the official `confluent-kafka` library in our local Python environment to communicate with the broker.

```python
%pip install confluent-kafka
```

---

## <span style="color: #FF6B6B;">Step 4:</span> Review the Source Files

Before running the script, open the `producer.py` and `orders.csv` files in your editor to see how they work.

- `orders.csv`: Contains 5 sample orders.
- `producer.py`: Reads the CSV row by row, extracts `order_id` as the message **Key**, formats the row as a JSON **Value**, and uses `producer.produce()` to send it to Kafka asynchronously.

---

## <span style="color: #FF6B6B;">Step 5:</span> Run the Python Producer

Run the script below. You should see delivery reports confirming exactly which partition and offset each message was saved to.

```python
!python producer.py
```

---

## <span style="color: #FF6B6B;">Step 6:</span> Verify Data with Console Consumer

Let's use the standard `kafka-console-consumer` to check what arrived inside the Kafka broker. Notice how the Key (`order_id`) and Value (JSON payload) are separated by a hyphen `-`.

```python
!docker exec kafka kafka-console-consumer \
  --bootstrap-server localhost:9092 \
  --topic orders \
  --from-beginning \
  --property print.key=true \
  --property key.separator=" - " \
  --timeout-ms 5000
```

Stop the cluster and remove volumes to clean up the disk space.

```python
!docker-compose -f ../../docker-compose.yml down -v
```
